````python
# ‰Ω†ÁöÑÁ®ãÂºè

import numpy as np
import matplotlib.pyplot as plt

# Âõ∫ÂÆö‰∫ÇÊï∏Á®ÆÂ≠ê‰ª•‰æøÁµêÊûúÂèØÈáçÁèæ
np.random.seed(768)

n = 200
x = np.random.normal(loc=2.0, scale=1.0, size=n)
y = x*0.5+np.random.normal(scale=0.8, size=n)
size_ver = np.abs(np.random.normal(loc=20, scale=10, size=n))
size_scale = (size_ver - size_ver.min()) / (size_ver.max() - size_ver.min() + 1e-9)  * 80 + 20

fig, ax = plt.subplots()

sc = ax.scatter(x, y, c=y, s=size_scale, alpha=0.75)
cbar = plt.colorbar(sc, ax=ax)
cbar.set_label("y-values")

ax.set_title("Scatter with Colormap")
ax.set_xlabel("x")
ax.set_ylabel("y")

plt.tight_layout
plt.savefig
plt.show()

````
